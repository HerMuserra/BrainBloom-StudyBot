```python
# ------------------------------
# BrainBloom Study Buddy
# ------------------------------

# Step 1: Load notes from a file
notes = {}
try:
    with open('notes.txt', 'r') as f:
        for line in f:
            if ":" in line:
                key, value = line.strip().split(":", 1)
                notes[key.strip()] = value.strip()
except FileNotFoundError:
    print("notes.txt not found! Make sure it's in the same folder as this notebook.")
    notes = {
        "AI": "Artificial Intelligence is the simulation of human intelligence in machines.",
        "Machine Learning": "Machine Learning is a subset of AI where machines learn from data.",
        "Python": "Python is a versatile programming language used for AI, web dev, and more."
    }

# Step 2: Define the bot function
def ask_bot(topic):
    topic = topic.lower()
    for key in notes:
        if key.lower() == topic:
            return notes[key]
    return "Hmmâ€¦ I don't know about that. Try another topic!"

# Step 3: Interactive loop
print("Welcome to BrainBloom Study Buddy! Type 'exit' to quit.")
while True:
    topic = input("Ask me about a topic: ")
    if topic.lower() == 'exit':
        print("Bye! Keep studying! ðŸ“š")
        break
    print(ask_bot(topic))
```
