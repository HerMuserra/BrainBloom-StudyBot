```python
import random

# Load quiz questions
quiz = {}
try:
    with open('flashquiz.txt', 'r') as f:
        for line in f:
            if ":" in line:
                q, a = line.strip().split(":", 1)
                quiz[q.strip()] = a.strip()
except FileNotFoundError:
    print("flashquiz.txt not found! Please add the file.")
    quiz = {
        "What is the capital of France?": "Paris",
        "Who developed Python?": "Guido van Rossum",
        "What is AI?": "Artificial Intelligence",
        "2 + 2 = ?": "4"
    }

# Quiz logic
score = 0
questions = list(quiz.keys())
random.shuffle(questions)

print("Welcome to FlashQuiz AI! Type 'exit' to quit anytime.\n")

for q in questions:
    ans = input(q + " ")
    if ans.lower() == 'exit':
        break
    if ans.strip().lower() == quiz[q].lower():
        print("✅ Correct!\n")
        score += 1
    else:
        print(f"❌ Wrong! Answer: {quiz[q]}\n")

print(f"Quiz Over! Your score: {score}/{len(questions)}")
```
